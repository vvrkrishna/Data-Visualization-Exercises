```python
import pandas as pd
%matplotlib inline
```

```python
names=['Bob','Jessica','Mary','John','Mel']
grades=[76,83,77,78,95]
GradeList=zip(names,grades)
df=pd.DataFrame(data=GradeList,columns=['names','grades'])
df.plot()
```

```python
#Code to plot customized matplotlib graph
import matplotlib.pyplot as plt
df.plot()
displayText="my annotation"
xloc=1
yloc=df['grades'].max()
xtext=8
ytext=-150
plt.annotate(displayText,xy=(xloc,yloc),arrowprops=dict(facecolor='black',shrink=0.05),xytext=(xtext,ytext),xycoords=('axes fraction','data'),textcoords='offset points')
```

```python
#Code to plot customized matplotlib graph
import matplotlib.pyplot as plt
df.plot(x = 'names', y = 'grades')
displayText="Wow!"
xloc=0
yloc=76
xtext=-150
ytext=-2
plt.annotate(displayText,xy=(xloc,yloc),arrowprops=dict(facecolor='black',shrink=0.05),xytext=(xtext,ytext),xycoords=('axes fraction','data'),textcoords='offset points')
```
