```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-whitegrid')
#Create empty figure
fig=plt.figure()
ax=plt.axes()
x=np.linspace(0,10,1000)
ax.plot(x,np.sin(x));
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
#Setting up x and y axis Range
plt.xlim(0,11)
plt.ylim(-2,2)
plt.axis('tight')

#Adding title to the chart
plt.title('Plotting data using sin and cos')
```
