```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
SalesMen=['Ahmed','Omar','Ali','Ziad','Salwa','Lila']
Mobile_Sales=[2540,1370,1320,2000,2100,2150]
TV_Sales=[2200,1900,2150,1850,1770,2000]
```

```python
df=pd.DataFrame()
df['Name']=SalesMen
df['Mobile_Sales']=Mobile_Sales
df['TV_Sales']=TV_Sales
df.set_index("Name",drop=True,inplace=True)
df
```

### Bar Plot of Sales Volume

```python
df.plot.bar(figsize=(20,10),rot=0).legend(bbox_to_anchor=(1,1))
plt.xlabel('Salesmen')
plt.xlabel('Sales')
plt.title('Sales Volume for two salesmen in \nJanuary and April 2017')
plt.show()
```

### Pie Chart of Item Sales

```python
df.plot.pie(subplots=True)
```

### Box plot of item sales

```python
df.plot.box()
```

### Area plot of item sales

```python
df.plot.area(figsize=(6,4)).legend(bbox_to_anchor=(1.5,1))
```

### Stacked Bar Plot of item Sales

```python
df.plot.bar(stacked=True,figsize=(20,10)).legend(bbox_to_anchor=(1.2,1))
```
